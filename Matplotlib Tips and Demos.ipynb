```
np.random.seed(362423)
A = np.random.uniform(size=(6,8))
```

```
print (A)
```

```
def plotEX(ax,**kw):
    ax.pcolormesh(A,**kw)
```

```
fig,(a1,a2,a3) = plt.subplots(1,3,figsize=(12,4),dpi=100)
plotEX(a1,cmap=plt.cm.Paired)
plotEX(a2,cmap=plt.cm.Set2)
plotEX(a3,cmap=plt.cm.Dark2)
 
 
```
