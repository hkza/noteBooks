```
import matplotlib.pyplot as plt
from matplotlib.path import Path
import matplotlib.patches as patches

verts = [
   (0., 0.),  # left, bottom
   (0., 1.),  # left, top
   (1., 1.),  # right, top
   (1., 0.),  # right, bottom
   (0., 0.),  # ignored
]

codes = [
    Path.MOVETO,
    Path.LINETO,
    Path.LINETO,
    Path.LINETO,
    Path.CLOSEPOLY,
]

path = Path(verts, codes)

fig, ax = plt.subplots()
patch = patches.PathPatch(path, facecolor='orange', lw=2)
ax.add_patch(patch)
ax.set_xlim(0, 2)
ax.set_ylim(0, 2)
plt.show()
```

```
from pylab import *

x = linspace(-3, 3, 51)
y = linspace(-2, 2, 41)
X, Y = meshgrid(x, y)

Z = (1 - X/2 + X**5 + Y**3) * exp(-X**2 - Y**2) # calcul du tableau des valeurs de Z

pcolor(X, Y, Z)

show()
```

```
from pylab import *

x = linspace(-3, 2, 9)
y = linspace(-2 ,3, 9)
X, Y = meshgrid(x, y)

C = angle(X + 1j *Y) 
print (X)

pcolormesh(X, Y, C, shading='gouraud', cmap=cm.hsv)# flat
colorbar()
xlabel('partie reelle')
ylabel('partie imaginaire')


show()
```

```
import random
fig = plt.figure(1, figsize=(12, 6))
co=['#a6cee3','#1f78b4','#b2df8a','#33a02c','#fb9a99','#e31a1c','#fdbf6f','#ff7f00','#cab2d6']


barWidth = 0.8
y1 = random.sample(range(1,44), 10)
y2=  random.sample(range(1,55), 10)
y3=[a+b for (a, b) in zip(y1, y2)] # multiplications terme à terme

print(y1,y3)
r = range(len(y1))

pyplot.bar(r, y1, width = barWidth, 
           color = ['b' for i in y1],
           edgecolor = ['w' for i in y1],
           linewidth = 0)

pyplot.bar(r, y2, width = barWidth, bottom =max(y1)+5 ,color=co[1] )

pyplot.bar(r, y3, width = barWidth, bottom =max(y1)+max(y2)+5  ,color=co[2]  )

alpha=[ chr(i) for i in  range (97,97+len(y1))]

pyplot.xticks(range(len(y1)), alpha)
```
