```
import pandas as pd
train=pd.read_csv("http://bit.ly/kaggletrain")
train.head()
```

```
train.Embarked.value_counts()
```

```
pd.get_dummies(train,columns=['Sex','Embarked'],drop_first=True)
```

```
E_dum=pd.get_dummies(train.Embarked,prefix='Embarked').iloc[:,1:]
```

```
train=pd.concat([train,E_dum],axis=1)
train.head()
```
